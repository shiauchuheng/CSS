```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
#import scrapy
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
#1.because of there is no universial structrue, we need to write one crawler for each webpage
#2.even by doing this,some webpages aare difficult to get the bios and names corresponding
url="https://www.iscb.org/recomb-regsysgen2013-program/recomb-regsysgen2013-keynote-spk-ab-bio"
response=urlopen(url)
soup=BeautifulSoup(response.read(),'html.parser',from_encoding='utf-8')
nodes=soup.find_all('p')
str1=""
for node in nodes:
    
    temp_text=node.get_text()
    temp_text=temp_text.split("................")
    for text in temp_text:
#         print(text)
#         print("################################################")
         p=re.findall(r'Biography:(.*)',text)
         if(len(p)>0):
             print(p)
             print("**********************************************")
    
   
div_tag=soup.find_all("div",{"class":"item-page"})
for tag in div_tag:
    a_tags=tag.find_all("a")
    for a in a_tags:
        print(a.get_text())


```
